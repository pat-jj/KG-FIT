```python
entity2label = {}

files = ['test', 'train', 'valid']

for file in files:
    with open(file + '.txt', 'r') as f:
        lines_text = f.readlines()
    with open(f'../{file}.txt', 'r') as f:
        lines_id = f.readlines()
        
    for i in range(len(lines_text)):
        head_text, relation, tail_text = lines_text[i].strip().split('\t')
        head_id, relation, tail_id = lines_id[i].strip().split('\t')
        if head_id not in entity2label:
            entity2label[head_id] = head_text
        else:
            if entity2label[head_id] != head_text:
                print(f'Error: {head_id} {entity2label[head_id]} {head_text}')
        if tail_id not in entity2label:
            entity2label[tail_id] = tail_text
        else:
            if entity2label[tail_id] != tail_text:
                print(f'Error: {tail_id} {entity2label[tail_id]} {tail_text}')
    
```
